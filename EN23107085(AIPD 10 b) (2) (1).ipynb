## ASSignment No:3
3. Fork an open-source AI project, create a feature branch, submit a pull
request, and address conflicts during the review process.

```python
STEP 1: Open Original Open-Source AI Project
Original Repository:
>> https://github.com/Jamesken23/Mechine-Learning.git
Open the repository in browser.
    
>>Click on the "Fork" button (top right corner).
Select your GitHub account and create fork.
    
This creates a copy in:
>>https://github.com/sakshinikam2005/Mechine_learning.git
------------------------------------------------------------

STEP 2: Configure Git (Only First Time)

## git config --global user.name "sakshinikam2005"
## git config --global user.email "sakshi.nikam.aids.2023@vpkbiet.org"

-------------------------------------------------------------
STEP 3: Clone Forked Repository

## git clone https://github.com/sakshinikam2005/Mechine_learning.git

-------------------------------------------------------------
STEP 4: Enter Project Directory
give fork repo name
## cd Mechine_learning
This moves into the project folder.
-------------------------------------------------------------
STEP 5: Create Feature Branch
##git checkout -b fork-practical
This creates and switches to a new branch named fork-practical.
-------------------------------------------------------------
STEP 6: Create New File

## nano fork_practical.txt
Add some content inside file.(Fork practical implementation)

This creates a new file for changes.
------------------------------------------------------------
STEP 7: Add Changes to Staging Area

## git add .
This stages all modified files.
-------------------------------------------------------------
STEP 8: Commit Changes
Command:

## git commit -m "Added fork practical file"
This saves changes locally.
------------------------------------------------------------
STEP 9: Push Branch to GitHub
Command:
## git push origin fork-practical

-------------------------------------------------------------
STEP 10: Create Pull Request
Open your forked repository on GitHub.
>> Click "Compare & pull request".
Add:
>> Title: Added fork practical file
>> Description: Fork practical implementation
>> Click "Create Pull Request".
This submits the PR to original repository.
-------------------------------------------------------------
STEP 11: Add Upstream Repository (To Handle Conflicts)
used fork name original
## git remote add upstream https://github.com/Jamesken23/Mechine-Learning.git
Check remotes:
## git remote -v
This connects original repository as upstream.
-------------------------------------------------------------
STEP 12: Fetch Upstream Changes
Command:
## git fetch upstream
This downloads latest changes from original repository.
------------------------------------------------------------
STEP 13: Merge Upstream Changes
Command:
## git merge upstream/main
If there is conflict, Git will show:
CONFLICT (content): Merge conflict in filename
```
